```python
import requests
def weather_api_request(weather_data):
    response = requests.get(
        (
            "https://api.open-meteo.com/v1/forecast"
        ),
        params = {
        "latitude": 47.374,
        "longitude": 8.540,
        "hourly": weather_data,
        "forecast_days": 16
    }
    )
    json_response = response.json()
    print(json_response)
    return json_response
```

```python
json_response = weather_api_request("rain")
hourly_data = json_response.get("hourly")
print(hourly_data)
```

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame(hourly_data)
print(df)

fig = px.line(df, x="time", y="rain")
fig.show()
```

```python
json_response = weather_api_request("temperature_2m")
hourly_data = json_response.get("hourly")
print(hourly_data)
```

```python
df = pd.DataFrame(hourly_data)
print(df)
fig = px.scatter(df, x="time", y="temperature_2m")
fig.show()
```
