```python
import requests


response = requests.get(
    (
        "https://api.open-meteo.com/v1/forecast"
    ),
    params = {
	"latitude": 52.52,
	"longitude": 13.41,
	"current": "temperature_2m",
	"hourly": "temperature_2m",
	"forecast_days": 16
}
)
json_response = response.json()
print(json_response)
```

```python
hourly_data = json_response.get("hourly")
print(hourly_data)
```
